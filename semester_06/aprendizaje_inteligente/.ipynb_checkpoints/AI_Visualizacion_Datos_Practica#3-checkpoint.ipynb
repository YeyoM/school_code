### Practica 3. Visualizacion de Datos

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Datos

```python
x = np.linspace(0, 10, 100)
print(x)
```

#### Ejemplos de graficos de lineas

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

#### Paneles de graficos

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

#### Estilo OOP

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

#### Usando colores :)

```python
plt.plot(x, np.sin(x - 0), color='blue')
plt.plot(x, np.sin(x - 1), color='g')
plt.plot(x, np.sin(x - 2), color='0.75')
# Tambien se pueden usar #fff (hex), (0,0,0) (rgb)
```

#### Tipos de lineas

```python
plt.plot(x, x + 0, linestyle="dashed")
plt.plot(x, x + 1, linestyle="solid")
plt.plot(x, x + 2, linestyle="dotted")
```

#### Cambiando limites

```python
plt.plot(x, np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5, 1.5)
```

#### Leyendas

```python
plt.plot(x, np.sin(x), '-g', label='Seno(x)')
plt.plot(x, np.cos(x), ':b', label='Coseno(x)')
plt.legend(loc='upper right')
```

#### Ejemplos con los datos de la flor de Iris

```python
import numpy as np
import pandas as pd
df = pd.read_csv('iris.csv', delimiter=';', decimal='.')
print(df.shape)
print(df.head())
```

#### Eliminar variable de categoria, dejar las variables predictoras en x (setosa y versicolor)

```python
x = df.iloc[0:100, [0, 2]].values
```

```python
plt.scatter(x[:50, 0], x[:50, 1],
            color="red", marker="o", label="setosa")
plt.scatter(x[50:100, 0], x[50:100, 1],
            color="blue", marker="x", label="versicolor")

plt.xlabel('sepal length')
plt.ylabel('petal length')

plt.legend(loc='upper left')
```

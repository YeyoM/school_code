# Modelo Discriminante Gausiano

#### Un poco de contexto

El Modelo Discriminante Gaussiano (GDA) es un algoritmo de aprendizaje supervisado utilizado para la clasificación de datos. Se basa en la suposición de que los datos de cada clase se distribuyen según una distribución gaussiana (también conocida como distribución normal).

#### Problema a resolver... Flor de Iris

El Modelo Discriminante Gaussiano (GDA) es adecuado para problemas de clasificación donde las variables de entrada son continuas y siguen una distribución gaussiana. En este ejemplo, se utilizará el conjunto de datos de Iris para crear un clasificador de flores utilizando el modelo GDA. El conjunto de datos tiene tres clases: setosa, versicolor y virginica.

#### Las fórmulas que utiliza este modelo...


```python
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from scipy.stats import multivariate_normal

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal
```

```python
def fit(x_train, y_train):
    m = y_train.shape[0] # Number of training example
    #Reshapeing the training set
    x_train = x_train.reshape(m, -1)
    input_feature = x_train.shape[1] # Number of input feature. In our case it's 4
    class_label = len(np.unique(y_train.reshape(-1))) # Number of class. In our case its 3.
    
    # Start everything with zero first.
    # Mean for each class. Each row contains an individual class. And each of the class input is 4 dimenstional
    mu = np.zeros((class_label, input_feature))
    # Each row will conatain the covariance matrix of each class.
    # The covariance matrix is a square symettric matrix.
    # It indicates how each of the input feature varies with each other.
    sigma = np.zeros((class_label, input_feature, input_feature))
    # Prior probability of each class.
    # Its the measure of knowing the likelihood of any class before seeing the input data.
    phi = np.zeros(class_label)

    for label in range(class_label):
        # Seperate all the training data for a single class
        indices = (y_train == label)
        
        phi[label] = float(np.sum(indices)) / m
        mu[label] = np.mean(x_train[indices, :], axis=0)
        # Instead of writting the equation we used numpy covariance function. 
        sigma[label] = np.cov(x_train[indices, :], rowvar=0)
        # Plot the Gaussian distribution for each class
    
    return phi, mu, sigma
```

```python
def plot_gaussian_distributions(mu, sigma, class_labels):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    
    # Assuming the Iris dataset has 4 features, we'll plot the first two for simplicity
    # You can extend this to plot all features if needed
    for i, label in enumerate(class_labels):
        # Create a grid of points
        x = np.linspace(mu[i, 0] - 3*np.sqrt(sigma[i, 0, 0]), mu[i, 0] + 3*np.sqrt(sigma[i, 0, 0]), 100)
        y = np.linspace(mu[i, 1] - 3*np.sqrt(sigma[i, 1, 1]), mu[i, 1] + 3*np.sqrt(sigma[i, 1, 1]), 100)
        X, Y = np.meshgrid(x, y)
        
        # Compute the multivariate normal distribution
        Z = multivariate_normal.pdf(np.dstack((X, Y)), mean=mu[i, :2], cov=sigma[i, :2, :2])
        
        # Plot the distribution
        ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, label=label)
    
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_zlabel('Probability Density')
    ax.legend()
    plt.show()
```

```python
def plot_gaussian_distributions_2d(mu, sigma, class_labels):
    # Assuming the Iris dataset has 4 features, we'll plot the first two for simplicity
    # You can extend this to plot all features if needed
    for i, label in enumerate(class_labels):
        # Create a grid of points
        x = np.linspace(mu[i, 0] - 3*np.sqrt(sigma[i, 0, 0]), mu[i, 0] + 3*np.sqrt(sigma[i, 0, 0]), 100)
        y = np.linspace(mu[i, 1] - 3*np.sqrt(sigma[i, 1, 1]), mu[i, 1] + 3*np.sqrt(sigma[i, 1, 1]), 100)
        X, Y = np.meshgrid(x, y)
        
        # Compute the multivariate normal distribution
        Z = multivariate_normal.pdf(np.dstack((X, Y)), mean=mu[i, :2], cov=sigma[i, :2, :2])
        
        # Plot the distribution
        plt.contourf(X, Y, Z, alpha=0.5, label=label)
    
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()
```

```python
def gaussian_pdf(x, mean, std_dev):
    return (1.0 / (np.sqrt(2 * np.pi) * std_dev)) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

def plot_gaussian_bell_curve_feature(mu, sigma, class_labels, feature_index):
    for i, label in enumerate(class_labels):
        # Generate x values for the specified feature
        x = np.linspace(mu[i, feature_index] - 3*np.sqrt(sigma[i, feature_index, feature_index]), 
                        mu[i, feature_index] + 3*np.sqrt(sigma[i, feature_index, feature_index]), 100)
        y = gaussian_pdf(x, mu[i, feature_index], np.sqrt(sigma[i, feature_index, feature_index]))
        
        # Plot the bell curve for the specified feature
        plt.plot(x, y, label=label)
    
    plt.xlabel(f'Feature {feature_index + 1}')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()
```

1. Cálculo de la probabilidad de clase (phi):
- El código comienza calculando la probabilidad de cada clase en el conjunto de datos de entrenamiento. Esto se hace contando cuántas veces aparece cada clase en el conjunto de datos y dividiendo ese número por el total de ejemplos de entrenamiento (m).
- En la implementación, phi es un vector que contiene las probabilidades de cada clase.

2. Cálculo de las medias y matrices de covarianza para cada clase (mu y sigma):
- Luego, se calculan las medias y matrices de covarianza para cada clase. Estos parámetros se utilizan para modelar la distribución de datos de cada clase como una distribución gaussiana multivariante.
- Mu es una matriz donde cada fila representa la media de las características para una clase específica.
- Sigma es una matriz tridimensional donde cada matriz de dos dimensiones (input_feature x input_feature) representa la matriz de covarianza para una clase específica.

3. Iteración sobre las clases:
- El código itera sobre cada clase y calcula phi, mu y sigma para esa clase.

```python
def predict(x_tests, phi, mu, sigma):
    # flatten the training data
    x_tests = x_tests.reshape(x_tests.shape[0], -1)
    class_label = mu.shape[0] # Number of label we have in our case it's k = 2
    scores = np.zeros((x_tests.shape[0], class_label))  # Initially we set the each class probability to zero.
    for label in range(class_label): # We will calculate the probability for each of the class.
        # normal_distribution_prob.logpdf Will give us the log value of the PDF that we just mentioned above.
        normal_distribution_prob = multivariate_normal(mean=mu[label], cov=sigma[label])
        # x_test can have multiple test data we will calculate the probability of each of the test data
        for i, x_test in enumerate(x_tests):
            #print(x_test)
            #print(np.log(phi[label]) + normal_distribution_prob.logpdf(x_test))
            scores[i, label] = np.log(phi[label]) + normal_distribution_prob.logpdf(x_test)
    predictions = np.argmax(scores, axis=1)
    return predictions
```

Este código toma como entrada un conjunto de datos de prueba (x_tests) junto con los parámetros del modelo GDA (phi, mu y sigma) y devuelve las predicciones para cada instancia de prueba. Utiliza la distribución gaussiana multivariante para calcular las probabilidades de cada clase para cada instancia de prueba y luego selecciona la clase con la probabilidad más alta como la predicción final.







```python
# Cargamos el dataset 
data = load_iris()

# Filtramos para solo tener las clases de setosa y versicolor
setosa_indices = np.where(data.target == 0)[0]
versicolor_indices = np.where(data.target == 1)[0]
selected_indices = np.concatenate((setosa_indices, versicolor_indices))
data.data = data.data[selected_indices]
data.target = data.target[selected_indices]

# Dividimos los datos para entrenar y hacer tests
x_train, x_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)

# Hacemos el fit del modelo 
phi, mu, sigma = fit(x_train, y_train)

class_labels = ['Iris-setosa', 'Iris-versicolor']

# Call the function with the means, covariances, and class labels
#plot_gaussian_distributions(mu, sigma, class_labels)
#plot_gaussian_distributions_2d(mu, sigma, class_labels)

# Call the function with the means, covariances, and class labels for feature 1
plot_gaussian_bell_curve_feature(mu, sigma, class_labels, 0)

# Call the function with the means, covariances, and class labels for feature 2
plot_gaussian_bell_curve_feature(mu, sigma, class_labels, 1)

y_predict = predict(x_test, phi, mu, sigma)
score = f1_score(y_test, y_predict, average="weighted")
print("f1 score of our model: ", score)

# Compare this model with scikitlearn LinearDiscriminatorAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)
y_predict_sk = lda.predict(x_test)
print("f1 score of scikit-learn model is: ", f1_score(y_test, y_predict_sk, average="weighted"))
```

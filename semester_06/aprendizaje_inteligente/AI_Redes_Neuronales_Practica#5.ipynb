#### Práctica No. 5. Redes Neuronales

```python
import  os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier # Multi Layer Perceptron Classifier
from sklearn.metrics import confusion_matrix
from pandas import DataFrame
```

```python
datos = pd.read_csv('iris.csv', delimiter=';', decimal='.')
print(datos.shape)
print(datos.head())
```

#### Dejamos las variables predictoras en x

```python
x = datos.iloc[:, :4]
print(x.head())
```

#### Dejar la variable a predecir en y

```python
y = datos.iloc[:, 4:5]
print(y.head())
```

#### Separar datos 70% entrenamiento, 30% testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, random_state = 0)
```

#### Mediante el constructor inicializamos la red neuronal
nota. Por defecto se usan 100 nodos en cada capa oculta, "solver" es el metodo de optimización (este nos ayudará para el gradiente descendente, hay muchos, como adam, stocastic, ...)

```python
red = MLPClassifier()
print(red)
```

#### Entrenando al modelo

Se logra llamando al método fit.

```python
red.fit(x_train, y_train.iloc[:,0].values)
```

#### Imprimiendo las predicciones en testing

```python
print("Las predicciones en testing son: {}".format(red.predict(x_test)))
```

#### Calculando los índices de calidad de la predicción

```python
# Funcion
def indices_general(MC, nombres = None):
    precision_global = np.sum(MC.diagonal()) / np.sum(MC)
    error_global = 1 - precision_global
    precision_categoria = pd.DataFrame(MC.diagonal() / np.sum(MC, axis = 1)).T
    if nombres != None:
        precision_categoria.columns = nombres
    return {"Matriz de confusion": MC,
                  "Precision Global": precision_global,
                  "Error Global": error_global,
                  "Precision por categoria": precision_categoria
           }
```

```python
prediccion = red.predict(x_test)
mc = confusion_matrix(y_test, prediccion)
indices = indices_general(mc, list(np.unique(y)))
for i in indices:
    print("\n%s:\n%s"%(i, str(indices[i])))
```
